<h3 style="color: maroon"><b>OBJECTIVE</b></h3>
In today’s digital world, individuals spend a significant portion of their daily lives using technological devices such as smartphones, laptops, and gaming consoles. While these technologies improve productivity and connectivity, excessive screen time may negatively impact sleep patterns, stress levels, physical activity, and overall mental wellbeing.However, there is limited data-driven understanding of how different types of gadget usage influence mental health outcomes and which lifestyle factors contribute most to stress, sleep quality, and productivity.This project aims to analyze real-world behavioral data to identify patterns, trends, and relationships between technology usage and mental wellbeing using exploratory data analysis techniques.

<h3 style="color: maroon"><b>DATA SET DESCRIPTION</b></h3>
The dataset contains information collected from 1,000 individuals across different countries, focused on understanding the relationship between daily technology usage and mental wellbeing. It includes demographic details, tech usage patterns, lifestyle habits, and mental health indicators.

Column Descriptions:

* User.ID - A unique identifier for each participant in the dataset.
* Age - Age of the participant, ranging from 16 to 64 years old.
* Gender - Gender of the participant. Possible values: Male, Female, or Other.
* Country - The country where the participant lives. Includes countries like USA, India, UK, Germany, Canada, and Australia.
* Device.Type - The primary device used most frequently by the participant. Possible values: Phone, Laptop, Tablet, or Desktop.
* Daily.Screen.Time - Total screen time (in hours) spent per day across all devices.
* Social.Media.Hours - Average number of hours per day spent on social media platforms (e.g., Instagram, TikTok, Twitter).
* Work.Screen.Hours - Number of screen hours per day dedicated to work or academic tasks.
* Gaming.Hours - Number of hours spent on gaming per day.
* Sleep.Hours - Average sleep duration in hours per day.
* Sleep.Quality - Self-reported sleep quality rated on a scale from 1 (very poor) to 10 (excellent).
* Stress.Level - Self-reported daily stress level rated on a scale from 1 (very low stress) to 10 (extremely high stress).
* Physical.Activity.Min - Total number of minutes of physical activity performed per day.
* Productivity.Score - Self-rated productivity score on a scale from 1 (very low) to 10 (very high), reflecting how effective the person feels throughout the day.
* Mental.Health.Status - The participant's overall mental health status. Categorised as Healthy: No known mental health issues, Moderate Issues: Some stress, anxiety, or occasional low mood, Chronic Issues: Ongoing or clinically significant mental health concerns.

### <span style="color:maroon">Creating Derived Columns:</span>

**a) Age Group** :Creating a new column AgeGroup based on the existing column Age.

Age Range	    Segment Name


16–19	    Adolescents          	   -School phase, high digital exposure

20–29	    Young Adult                -College / early career

30–39	    Early Mid-life             - Career building

40–49	    Mid-life                   -Stable career & family phase

50-64	    Older Adults               - Lower tech adoption tendency

```python
number=[16,20,30,40,50,65]
name=['Adolescents','Young Adult','Early Mid-life','Mid-life','Older Adults']
df['AgeGroup']=pd.cut(df['Age'],bins=number,labels=name,right=False)
```

**b) Total Screen Time:** Calculating Total Screen Time by combining the available screen time columns -Social.Media.Hours,Work.Screen.Hours and Gaming.Hours

```python
df['TotalScreenTime']=df['Social.Media.Hours']+df['Work.Screen.Hours']+df['Gaming.Hours']
```

**c) Leisure Screen Time:** Calculating total  time spend on screen for non work purposes.

```python
df['LeisureScreenTime']=df['Social.Media.Hours']+df['Gaming.Hours']
```

**d) Physical Activity Level:** Creating PhysicalActivityLevel column based on the Physical.Activity.Min as

'Sedentary':0-29min,

'Moderate':30-59min,

'Active':60-119min.**

```python
activetime=[0,30,60,120]
activetype=['Sedentary','Moderate','Active']
df['PhysicalActivityLevel']=pd.cut(df['Physical.Activity.Min'],bins=activetime,labels=activetype,right=False)
```

**e) Screen Time Group:**  Creating ScreenTimeGroup column based on the Total Screen Time as


```python
df['ScreenTimeGroup'] = pd.cut(df['TotalScreenTime'],
                               bins=[0,2,4,6,8,12],
                               labels=['0-2','2-4','4-6','6-8','8+'])
```
